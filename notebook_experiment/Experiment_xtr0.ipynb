```
import sys
import os
sys.path.append('/content/drive/MyDrive/Kernel_Challenge/')
from tqdm.notebook import tqdm as tqdm
from kernel import *
from classifiers import *
from utils.utils import return_training_datasets, split_train_val
```

```
all_x, all_y = return_training_datasets('/content/drive/MyDrive/Kernel_Challenge/kernel_data')
```

# Xtr0

```
X, y = all_x[0], all_y[0]
x_train, x_val, y_train, y_val = split_train_val(X,y)
```

## Spectrum Kernel Grid Search

```
k_candidates = [5, 6, 7, 8, 9, 10, 11, 12]
matrix_ready = {k: False for k in k_candidates}
```

### SVM

```
C_candidates = [1, 0.5, 0.1, 0.07, 0.03, 0.01]
```

```
summary_SKSVM = dict()
for k in tqdm(k_candidates):
  for C in tqdm(C_candidates):
    Kernel = SpectrumKernel(k=k, add_inverse=True)
    clf = SVM(Kernel, C=C, save_dir='/content/drive/MyDrive/Kernel_Challenge/kernel_data/', version="Xtr0")
    path_2_save = os.path.join(clf.save_dir, clf.kernel.name + f"_{clf.version}" + ".pkl")
    if matrix_ready[k]:
      kernel = Kernel.load_kernel(path_2_save)
      clf.kernel = kernel
      clf._fit(y_train.Bound)
    else:
      clf.fit(x_train, y_train.Bound)
      matrix_ready[k] = True
    y_pred = clf.predict(x_val)
    score = sum(y_pred==y_val.Bound)/len(y_pred)
    summary_SKSVM[f"C={C} and k={k} : Accuracy on validation"] = score
```

```
summary_SKVM
```

### Kernel Logisitic Regression

```
k_candidates = [5, 6, 7, 8, 9, 10, 11, 12]
matrix_ready = {k: True for k in k_candidates}
```

```
lambda_candidates = [0.1, 0.05, 0.01, 0.005, 0.001, 0.0005, 0.0001]
```

```
summary_SKLR = dict()
for k in tqdm(k_candidates):
  for lmbd in tqdm(lambda_candidates):
    Kernel = SpectrumKernel(k=k, add_inverse=True)
    clf = KernelLogisticRegression(Kernel, lambda_=lmbd, save_dir='/content/drive/MyDrive/Kernel_Challenge/kernel_data/', max_iter=10000, version="Xtr0")
    path_2_save = os.path.join(clf.save_dir, clf.kernel.name + f"_{clf.version}" + ".pkl")
    if matrix_ready[k]:
      kernel = Kernel.load_kernel(path_2_save)
      clf.kernel = kernel
      clf._fit(y_train.Bound)
    else:
      clf.fit(x_train, y_train.Bound)
      matrix_ready[k] = True
    y_pred = clf.predict(x_val)
    score = sum(y_pred==y_val.Bound)/len(y_pred)
    summary_SKLR[f"lambda={lmbd} and k={k} : Accuracy on validation"] = score
```

```
summary_SKLR
```

Summary : \
lambda=0.005 and k=11 : Accuracy on validation': 0.675 \

## Substring Kernel Grid Search

```
k_candidates = [6, 7, 8, 9, 10, 11]
m_candidates = [1, 2, 3]
matrix_ready = {(k,m): False for k,m in zip(k_candidates, m_candidates)}
```

### SVM

```
C_candidates = [1, 0.1, 0.01]
```

```
summary_SubKSVM = dict()
for k in tqdm(k_candidates):
  for m in tqdm(m_candidates):
    for C in tqdm(C_candidates):
      Kernel = SubstringKernel(k=k, delta=1, m=m)
      clf = SVM(Kernel, C=C, save_dir='/content/drive/MyDrive/Kernel_Challenge/kernel_data/', version="Xtr0")
      path_2_save = os.path.join(clf.save_dir, clf.kernel.name + f"_{clf.version}" + ".pkl")
      if matrix_ready[(k,m)]:
        kernel = Kernel.load_kernel(path_2_save)
        clf.kernel = kernel
        clf._fit(y_train.Bound)
      else:
        clf.fit(x_train, y_train.Bound)
        matrix_ready[k] = True
      y_pred = clf.predict(x_val)
      score = sum(y_pred==y_val.Bound)/len(y_pred)
      summary_SubKSVM[f"C={C} and k={k} and m={m} : Accuracy on validation"] = score
```

Summary : \
C=0.1 and k=10 : Accuracy on validation': 0.67 \
C=0.01 and k=11 : Accuracy on validation': 0.665 \
